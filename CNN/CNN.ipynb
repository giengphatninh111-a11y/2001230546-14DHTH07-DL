```python
import tensorflow as tf
import numpy as np

print("=== PHẦN 1: THAO TÁC TENSOR ===")

# 1. Tạo một tensor kích thước [2, 3] với các phần tử được gán 0
t1 = tf.zeros([2, 3])
print(f"Câu 1:\n{t1}\n")

# 2. Cho X = [[1, 2, 3], [4, 5, 6]], tạo tensor có kích thước và dtype như X gán 0
X_init = [[1, 2, 3], [4, 5, 6]]
t2 = tf.zeros_like(X_init)
print(f"Câu 2:\n{t2}\n")

# 3. Tạo một tensor kích thước [3, 2] với các phần tử được gán 5
t3 = tf.fill([3, 2], 5)
print(f"Câu 3:\n{t3}\n")

# 4. Tạo một tensor 1-chiều với 50 phần tử chẵn trong khoảng [5, 10]
# Lưu ý: Các số chẵn từ 5 đến 10 là 6, 8, 10. Ta dùng tf.repeat hoặc lặp để sinh đủ 50 phần tử.
even_numbers = [6, 8, 10]
t4 = tf.constant(np.resize(even_numbers, 50))
print(f"Câu 4 (Kích thước: {t4.shape}):\n{t4}\n")

# 5. Tạo một tensor 1-chiều có giá trị như [10, 12, 14, ..., 100]
t5 = tf.range(start=10, limit=102, delta=2)
print(f"Câu 5:\n{t5}\n")

# 6. Tạo một tensor [3, 2] ngẫu nhiên theo phân phối normal (mean=0, std=2)
t6 = tf.random.normal(shape=[3, 2], mean=0.0, stddev=2.0)
print(f"Câu 6:\n{t6}\n")

# 7. Tạo một tensor [3, 2] ngẫu nhiên theo phân phối uniform trong phạm vi [0, 2]
t7 = tf.random.uniform(shape=[3, 2], minval=0.0, maxval=2.0)
print(f"Câu 7:\n{t7}\n")

# 8. Cho tensor X = [[1, 2], [3, 4], [5, 6]]. Hãy shuffle X theo chiều thứ nhất
X_shuffle = tf.constant([[1, 2], [3, 4], [5, 6]])
t8 = tf.random.shuffle(X_shuffle)
print(f"Câu 8:\n{t8}\n")

# 9. Tạo ngẫu nhiên tensor [10, 10, 3]. Cắt thành tensor kích thước [5, 5, 3] bằng random_crop
large_tensor = tf.random.uniform(shape=[10, 10, 3])
t9 = tf.image.random_crop(large_tensor, size=[5, 5, 3])
print(f"Câu 9 (Kích thước sau cắt: {t9.shape}):\n{t9}\n")

# 10. Tạo ngẫu nhiên x, y (1-chiều, 5 phần tử). Trả về x + y nếu x < y, và x - y nếu x > y
x_10 = tf.random.uniform(shape=[5], minval=1, maxval=10, dtype=tf.int32)
y_10 = tf.random.uniform(shape=[5], minval=1, maxval=10, dtype=tf.int32)
# Sử dụng tf.where(điều kiện, giá trị_khi_đúng, giá trị_khi_sai)
t10 = tf.where(x_10 < y_10, x_10 + y_10, x_10 - y_10)
print(f"Câu 10:\nx = {x_10.numpy()}\ny = {y_10.numpy()}\nKết quả = {t10.numpy()}\n")

# 11. x = (3, 2) trong [5, 10], y = (2, 3) trong [15, 20]. Giữ đường chéo x và thực hiện x + y
x_11 = tf.random.uniform(shape=[3, 2], minval=5, maxval=11, dtype=tf.int32)
y_11 = tf.random.uniform(shape=[2, 3], minval=15, maxval=21, dtype=tf.int32)

# Lấy các phần tử đường chéo của x (với ma trận 3x2 đường chéo gồm x[0,0] và x[1,1])
diag_x = tf.linalg.tensor_diag_part(x_11)

# Để thực hiện x + y (phép cộng ma trận), hai ma trận phải cùng kích thước.
# Ta thực hiện chuyển vị y (Transpose) từ (2, 3) về (3, 2) để cộng với x.
y_11_transposed = tf.transpose(y_11)
sum_xy = x_11 + y_11_transposed
print(f"Câu 11:\nx = \n{x_11.numpy()}\ny (đã chuyển vị sang 3x2) = \n{y_11_transposed.numpy()}\nx + y = \n{sum_xy.numpy()}\nĐường chéo của x = {diag_x.numpy()}\n")
```

```python
print("=== PHẦN 2: XÂY DỰNG MODEL VÀ GRAPH MODE ===")

# =====================================================================
# 12. Hàm tạo 2 tensor ngẫu nhiên (3,2,2) và tính cộng, trừ, nhân, chia trong Graph
# =====================================================================
@tf.function
def graph_math_operations():
    # Khởi tạo ngẫu nhiên
    x = tf.random.uniform(shape=(3, 2, 2), minval=1, maxval=10)
    y = tf.random.uniform(shape=(3, 2, 2), minval=1, maxval=10)

    add_res = x + y
    sub_res = x - y
    mul_res = x * y
    div_res = x / y
    return x, y, add_res, sub_res, mul_res, div_res

x_gen, y_gen, r_add, r_sub, r_mul, r_div = graph_math_operations()
print(f"Câu 12: Đã chạy các phép toán cơ bản trong Graph Mode.")


# =====================================================================
# 13 + 14. Xây dựng mô hình tính Tổng, Hiệu và Tích hai tensor x, y
# =====================================================================
class BasicMathModel(tf.Module):
    @tf.function
    def __call__(self, x, y):
        tong = tf.add(x, y)
        hieu = tf.subtract(x, y)
        tich = tf.multiply(x, y)
        return tong, hieu, tich

math_model = BasicMathModel()
res_tong, res_hieu, res_tich = math_model(tf.constant([5.0]), tf.constant([3.0]))
print(f"Câu 13+14:\n- Tổng: {res_tong.numpy()}\n- Hiệu: {res_hieu.numpy()}\n- Tích: {res_tich.numpy()}")


# =====================================================================
# 15. Tính chu vi hình chữ nhật P = (a + b) * 2
# =====================================================================
class RectangleModel(tf.Module):
    @tf.function
    def compute_perimeter(self, a, b):
        return tf.multiply(tf.add(a, b), 2.0)

rect_model = RectangleModel()
print(f"Câu 15 (Chu vi HCN a=4, b=7): {rect_model.compute_perimeter(4.0, 7.0).numpy()}")


# =====================================================================
# 16. Tính diện tích toàn phần hình hộp chữ nhật
# =====================================================================
class BoxAreaModel(tf.Module):
    @tf.function
    def compute_areas(self, a, b, c):
        # S_xq = (a + b) * c * 2
        s_xq = tf.multiply(tf.multiply(tf.add(a, b), c), 2.0)
        # S_tp = (a * b) * 2 + S_xq
        s_tp = tf.add(tf.multiply(tf.multiply(a, b), 2.0), s_xq)
        return s_xq, s_tp

box_model = BoxAreaModel()
sxq, stp = box_model.compute_areas(3.0, 4.0, 5.0)
print(f"Câu 16 (Hình hộp 3x4x5):\n- S xung quanh: {sxq.numpy()}\n- S toàn phần: {stp.numpy()}")


# =====================================================================
# 17. Chứng minh hai cạnh bên hình thang song song
# Điều kiện: AD // BC <=> AD = BC và AB = CD
# =====================================================================
class TrapezoidParallelModel(tf.Module):
    @tf.function
    def prove_parallel(self, len_AD, len_BC, len_AB, len_CD):
        # Kiểm tra đồng thời hai điều kiện bằng tf.logical_and
        cond1 = tf.equal(len_AD, len_BC)
        cond2 = tf.equal(len_AB, len_CD)
        is_parallel_trapezoid = tf.logical_and(cond1, cond2)
        return is_parallel_trapezoid

trap_model1 = TrapezoidParallelModel()
print(f"Câu 17 (Kiểm tra hình thang AD=5, BC=5, AB=4, CD=4): {trap_model1.prove_parallel(5.0, 5.0, 4.0, 4.0).numpy()}")


# =====================================================================
# 18. Chứng minh hai cạnh đáy hình thang bằng nhau
# Điều kiện: AB = CD <=> AD = BC và AD // BC
# =====================================================================
class TrapezoidBaseModel(tf.Module):
    @tf.function
    def prove_equal_bases(self, len_AD, len_BC, is_AD_parallel_BC):
        cond1 = tf.equal(len_AD, len_BC)
        # Kết hợp điều kiện bằng phép toán logic and
        is_equal_bases = tf.logical_and(cond1, is_AD_parallel_BC)
        return is_equal_bases

trap_model2 = TrapezoidBaseModel()
print(f"Câu 18 (AD=6, BC=6, đã song song): {trap_model2.prove_equal_bases(6.0, 6.0, tf.constant(True)).numpy()}")


# =====================================================================
# 19. Chứng minh NA = NC dựa trên tính chất đường trung bình tam giác
# Đầu vào: AM, MB, AN, NC.
# Hệ điều kiện: AM = MB và MN // BC => NA = NC
# =====================================================================
class TriangleMidpointModel(tf.Module):
    @tf.function
    def prove_midpoint_relation(self, len_AM, len_MB, is_MN_parallel_BC):
        # Điều kiện 1: M là trung điểm AB (AM = MB)
        is_M_midpoint = tf.equal(len_AM, len_MB)

        # Nếu M là trung điểm và MN // BC thì N bắt buộc phải là trung điểm AC (NA = NC)
        is_N_midpoint = tf.logical_and(is_M_midpoint, is_MN_parallel_BC)
        return is_N_midpoint

triangle_model = TriangleMidpointModel()
proven_result = triangle_model.prove_midpoint_relation(3.5, 3.5, tf.constant(True))
print(f"Câu 19 (Kết luận tính đúng đắn NA = NC): {proven_result.numpy()}")
```
